# Lab 3 - Exploratory Data Analysis (EDA)

## Dataset: Students Performance Dataset

### Objective
The goal of this project is to perform Exploratory Data Analysis (EDA) on a student performance dataset.
The analysis includes examining the data structure, handling missing values, removing duplicates,
creating visualizations, and identifying patterns and insights.

### Steps Performed:
1. Imported required libraries.
2. Loaded the dataset.
3. Checked data structure (info, shape, describe).
4. Checked and handled missing values.
5. Removed duplicate records.
6. Created visualizations.
7. Identified patterns and insights.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
sns.histplot(df['math score'])
plt.title("Math Score Distribution")
plt.show()
```

### Insight
The math scores show a normal-like distribution with most students scoring between 60 and 80.

```python
df.groupby('gender')[['math score','reading score','writing score']].mean().plot(kind='bar')
plt.title("Average Scores by Gender")
plt.show()
```

### Insight
Female students have higher average scores in reading and writing compared to male students.

```python
sns.scatterplot(x='reading score', y='writing score', data=df)
plt.title("Reading vs Writing Score")
plt.show()
```

### Insight
There is a strong positive correlation between reading and writing scores.
